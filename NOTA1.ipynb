## MVC (Model View Controller)

* Model = Define Data Structure
* View = Define Display (UI) ->
* Controller = Contiene Control Logic -> Realiza las validaciones, sanitizaciones de datos, manejo de errores, entre varias otras cosas que permiten que los datos que se mandan desde la Vista, lleguen de forma más ordenada y limpia su procesamiento en el backend que es donde se encuentra el modelo.

Se refiere a la separación de la interfaz de usuario (Vista) de la parte de datos y de negocio (Modelo), la cual está intermediada por un controlador (Controller).

### NAVEGACIÓN/MENÚ
La forma en cómo el usuario identificará la organizacón de tu información.
* Frases cortas
* Lenguaje simple y claro
* No revolver la info
* Las URLs deben ser intuitivas
* Mostrar jerarquía con categorías y subcategorías
* Términos y condiciones

### CATEGORÍAS Y SUBCATEGORÍAS
Para agrupar contenidos e información.
* Menú principal
+ Submenú

### PÁGINAS INDIVIDUALES / FINALES
A partir de aquí ya no hay más jerarquía y aquí descansa la mayor parte de la infromación.

### OTROS TIPOS DE ESTRUCTURAS DE SITIOS
1. MODELO SECUENCIAL
Modelo para dirigir al usuario por diferentes pasoso secuenciales con un orden de finalización.
2. MODELO DE MATRIZ
Permite una navegación general por todo el sitio con muchas opciones de info específica, la cual es compleja de agrupar en categorías, estos sitios por lo general utilizan links internos del mismo sitio para poder dirigirte a las secciones deseadas.
